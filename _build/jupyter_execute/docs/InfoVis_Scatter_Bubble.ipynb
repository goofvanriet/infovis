```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df = pd.read_csv('InfoVis_dataV7.csv', delimiter=';')

import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data=df, x='hdi',y='co2_per_capita')

plt.xlabel('HDI')
plt.ylabel('CO2 per capita')
plt.title('Scatterplot van CO2 per capita vs. HDI')

plt.show()
```

```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df = pd.read_csv('InfoVis_dataV7.csv', delimiter=';')

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
scatter = plt.scatter(x=df['gdp_per_capita'], y=df['hdi'], s=df['co2_per_capita']*1, alpha=0.8, c=df['co2_per_capita'], cmap='viridis_r', edgecolors='w', linewidth=0.5)

plt.colorbar(scatter, label='CO2-uitstoot per capita')

plt.xlabel('CO2 per capita')
plt.ylabel('HDI')
plt.title('Bubble chart van CO2 per capita vs. HDI (met grootte gebaseerd op GDP per capita)')

plt.show()
```
